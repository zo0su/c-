[해보기3] 기존 파일 내용에 추가로 데이터 입력하기

```python
# [실습7]
title = '책 제목'   # 변수 title에 '책 제목' 문자열 데이터 저장
writer = '저자'     # 변수 writer에 '저자' 문자열 데이터 저장
f = open('book.csv', 'w')   # 쓰기 모드(w)로 파일 열기
f.write(title+', ' + writer + '\n') # write() 함수로 파일에 내용 쓰기
f.close()
```

```python
title = '소나기'
writer = '황순원'
f = open('book.csv', 'a')
f.write(title + ', ' + writer + '\n')
f.close()
# 파일 전체 내용 출력하기
f = open('book.csv', 'r')   # 읽기 모드(r)로 파일 열기
data = f.read()             # read() 함수로 파일의 내용 전체 읽기
print(data)
```
