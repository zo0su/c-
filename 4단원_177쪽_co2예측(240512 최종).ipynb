#  4단계. 기계학습을 통한 모델 생성

학습 모델을 마들기 위해~~ 부터 다 그대로 쓰되, (1)의 코드를 아래 코드로 교체

```python
#학습할 속성이 될 X(독립변수)와 타겟(목표)속성이 될 y(종속변수)를 나누어 저장하기
X = df[['Engine Size(L)', 'Fuel Consumption Comb (L/100 km)']]
y = df['CO2 Emissions(g/km)']
```

(2) 훈련 데이터와 ~~~~ 도 그대로 작성

```python
#훈련 데이터와 테스트 데이터를 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

(3) 도 그대로 작성, 하지만 실행결과가 작으니 세로로 나열하지 않고 코드 오른쪽에 가로로 (2단) 넣어도 공간 효율적으로 구성 가능할 듯 합니다.

```python
#선형회귀를 이용해 모델 생성하기
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()  #선형회귀 모델 생성하기
linear_model.fit(X_train, y_train) #선형회귀 모델 학습하기
```

(4)

```python
#학습 모델을 이용해 X_test 데이터에 대한 y 기대값을 예측한 결과를 linear_predict 변수에 저장하기
linear_predict = linear_model.predict(X_test)
```

```python
#예측 결과가 들어있는 linear_predict 값 확인해보기
linear_predict
```

(5) 가상의 ~~

```python
#가상의 독립 변수 값을 넣어 예측값이 어떻게 되는지 확인해보기
# 예) 엔진 크기가 3, 평균 연료소비량이 7.5일 경우의 이산화 탄소 배출량 예측해보기
linear_model.predict([[3.0,7.5]])
```

# 5단계. 성능 평가

사이킷런에는 기계학습 모델의 성능 평가를 하기 위한 함수가 포함되어 있다. R2 score을 계산하기 위한 함수인 r2_score 함수를 사용한다.

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, linear_predict)
```

```python
linear_model.score(X_test, y_test)
```
