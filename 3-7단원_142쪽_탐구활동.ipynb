[탐구활동] 탐색 프로그램 작성하기

```python
import random
data = []                        # 비어 있는 data 리스트 생성

# ㉮
for i in range(1000):            # 데이터 집합의 크기를 1000으로 설정하고 1000개의 데이터 생성
    a = random.randint(1, 1000)  # 1이상 1000이하의 랜덤한 수를 변수 num에 대입
    data.append(a)               # 변수 num를 data 리스트에 추가

count1, count2 = 0, 0            # 각 정렬 알고리즘의 비교 횟수를 저장할 count1, count2에 0 대입
print(data)                      # 1000개의 수가 저장된 data 리스트 출력
search = int(input('검색할 데이터를 입력하세요: '))

def search1(a, x):               # search1 함수 정의
    global count1                # 비교 횟수를 저장할 변수 count1을 수정하기 위한 코드
    n = len(a)                   # num의 길이를 변수 n에 대입
    for i in range(0, n):
        count1 = count1 + 1
        if x == a[i]:
            return '순차 탐색할 경우 비교횟수는 ' + str(count1) + '회입니다.'
    return '존재하지 않습니다.'

def search2(a, x):               # search2 함수 정의
    a = sorted(a)                # ㉯
    global count2                # 비교 횟수를 저장할 변수 count2를 수정하기 위한 코드
    start = 0                    # 변수 start에 첫 번째 데이터의 인덱스 값을 대입
    end = len(a) - 1             # 변수 end에 마지막 데이터의 인덱스 값을 대입
    while start <= end :
        mid = (start + end) // 2
        count2 = count2 + 1
        if x == a[mid]:
            return '이진 탐색할 경우 비교횟수는 ' + str(count2) + '회입니다.'
        elif x > a[mid] :
            start = mid + 1
        else :
            end = mid - 1
    return '존재하지 않습니다.'

print(search1(data, search))
print(search2(data, search))
```
