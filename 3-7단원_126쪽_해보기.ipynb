[해보기3] 선택 정렬의 비교횟수 확인하기

1) 프로그램을 직접 실행하여 비교 횟수를 확인해 보자.

```python
# data 리스트에 정렬하고자 하는 값을 입력한다.
data = [3, 5, 6, 2, 1, 4]
def selection_sort(data):
    print('정렬 전 데이터', data)
    count = 0
    for i in range(len(data) - 1):
        min_idx = i
        for j in range(i + 1, len(data)):
            count = count + 1
            if data[j] < data[min_idx]:
                min_idx = j
        data[i], data[min_idx] = data[min_idx], data[i]
    print('정렬 후 데이터', data)
    print('총 비교횟수:', count)
selection_sort(data)
```
